```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://books.toscrape.com/catalogue/category/books/fiction_10/page-1.html"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")
# print(soup.prettify())
```

```python
details = soup.find_all("article", class_ = "product_pod")
# print(details)
```

```python
titles = []
prices = []
availability = []
ratings = []
current_page = 1

while current_page <= 4:
    url = f"https://books.toscrape.com/catalogue/category/books/fiction_10/page-{str(current_page)}.html"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    
    for detail in details:
        book_name = (detail.find("img").attrs["alt"])
        titles.append(book_name)

        book_price = detail.find("p", class_ = "price_color").text[1::]
        prices.append(book_price)

        book_availability = detail.find("p", class_ ="instock availability").text.strip()
        availability.append(book_availability)   
        
        book_rating = detail.find("p", class_ = "star-rating")["class"][1]
        ratings.append(book_rating)
    
    current_page +=1 
    
# print(titles)
# print(prices)
# print(availability)
# print(ratings)
```

```python
data = pd.DataFrame({"Book Name":titles, "Book Price":prices, "Availability":availability,"Rating":ratings})
data.index = data.index + 1
```
